### Comparing data with multiple plots

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))

for i in range(0,4):
    ax=fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[i]], label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[i]], label='Men')
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(major_cats[i])
    ax.tick_params(bottom=False,top=False,right=False,left=False)
    plt.legend(loc='upper right')

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
#plt.legend(loc='upper right')
plt.show()
```

## Color, Layout, and Annotations

### Color

### Setting Line Color Using RGB

```python
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    cb_dark_blue = (0/255, 107/255, 164/255)
    cb_orange = (255/255, 128/255,14/255)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(major_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

### Setting Line Width

```python
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], linewidth=3, c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], linewidth=3, c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(major_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

### Improve the Layout and Ordering

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], linewidth=3, c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], linewidth=3, c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)


plt.legend(loc='upper right')
plt.show()
```

### Replacing the Legend With Annotations

```python
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], linewidth=3, c=cb_dark_blue, label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], linewidth=3, c=cb_orange, label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp==0:
        ax.text(2005,90,"Men")
        ax.text(2005,7,"Women")
    elif sp==5:
        ax.text(2005,70,"Men")
        ax.text(2005,30,"Women")



    
plt.show()
```

### Creating Histograms In Seaborn

```python
import seaborn as sns    # seaborn is commonly imported as `sns`
import matplotlib.pyplot as plt
```

```python
sns.histplot(titanic['Fare'],kde=True,stat="percent")
plt.show()
```

```python
sns.distplot(titanic['Age'])
plt.show()
```

### Generating A Kernel Density Plot

```python
sns.kdeplot(titanic['Age'],shade=True)
plt.show()
```

### Modifying The Appearance Of The Plots

```python
#sns.set_style()
```

```python
#sns.despine()
```

```python
sns.set_style('white')
sns.kdeplot(titanic['Age'],shade=True)
sns.despine(left=True,bottom=True)
plt.show()
```

### Conditional Plots

```python
#Condition on unique values of the "survived" column
g=sns.FacetGrid(titanic,col="Survived",height=6)

#For each subset of values create a certain chart 
g.map(sns.kdeplot,"Age",shade=True)

plt.show()

```

```python
#Condition on unique values of the "survived" column
g=sns.FacetGrid(titanic,col="Survived",height=6)

#For each subset of values create a certain chart 
g.map(plt.hist,"Age")

plt.show()
```

```python
#Condition on unique values of the "survived" column
g=sns.FacetGrid(titanic,col='Survived',row="Pclass",height=3)

#For each subset of values create a certain chart 
g.map(sns.kdeplot,"Age",shade=True)

plt.show()
```

```python
#Condition on unique values of the "survived" column
g=sns.FacetGrid(titanic,col='Survived',row="Pclass",hue='Sex',height=3)

#For each subset of values create a certain chart 
g.map(sns.kdeplot,"Age",shade=True)

plt.show()
```

###  Adding A Legend

```python
#Condition on unique values of the "survived" column
g=sns.FacetGrid(titanic,col='Survived',row="Pclass",hue='Sex',height=3)

#For each subset of values create a certain chart 
g.map(sns.kdeplot,"Age",shade=True)
g.add_legend()
plt.show()
```
