## Formatting And Spacing

```python
first_twelve = unrate[:12]
second_twelve = unrate[12:24]

fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(first_twelve["DATE"], first_twelve["VALUE"])
ax2.plot(second_twelve["DATE"], second_twelve["VALUE"])

ax1.set_title("Monthly unemployment rate, 1948")
ax2.set_title("Monthly unemployment rate, 1949")

plt.show()
```

## Comparing Across More Years

```python
fig = plt.figure(figsize=(12, 12))

for i in range(5):
    ax = fig.add_subplot(5, 1, i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index : end_index]
    ax.plot(subset["DATE"], subset["VALUE"])
    
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

colors = ["red", "blue", "green", "orange", "black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index : end_index]
    ax.plot(subset["MONTH"], subset["VALUE"], c=colors[i])
    
plt.show()
```

## Adding A Legend

<p><img alt="Legend" src="https://s3.amazonaws.com/dq-content/legend.png"></p>

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

colors = ["red", "blue", "green", "orange", "black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index : end_index]
    ax.plot(subset["MONTH"], subset["VALUE"], c=colors[i], label=str(1948+i))

ax.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

colors = ["red", "blue", "green", "orange", "black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index : end_index]
    ax.plot(subset["MONTH"], subset["VALUE"], c=colors[i], label=str(1948+i))

ax.legend(loc="upper left")
ax.set_title("Monthly unemployment trends 1948 - 1952")
ax.set_xlabel("Months")
ax.set_ylabel("Unemployment rate, %")
plt.show()
```

## Gridlines

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

colors = ["red", "blue", "green", "orange", "black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index : end_index]
    ax.plot(subset["MONTH"], subset["VALUE"], c=colors[i], label=str(1948+i))

ax.legend(loc="upper left")
ax.set_title("Monthly unemployment trends 1948 - 1952")
ax.set_xlabel("Months")
ax.set_ylabel("Unemployment rate, %")

ax.minorticks_on()
ax.grid(which="both", axis="y")
plt.show()
```

# Bar plots

## Introduction to the data

```python
reviews = pd.read_csv("data/fandango_scores.csv")
```

```python
cols = ["FILM", "RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
```

```python
norm_reviews = reviews[cols]
```

```python
norm_reviews.head(10)
```

## Bar Plots

<p><img alt="Vertical Bar Plot" src="images/vertical_bar_plot.png"></p>

### Creating Bars

<p><img alt="Matplotlib Barplot Positioning" src="images/matplotlib_barplot_positioning.png"></p>

```python
review_cols = ["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]

movie = norm_reviews.loc[0, :]
print(movie)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar([1,2,3,4,5], movie[review_cols], width=-0.3, align="edge")

plt.show()
```

### Aligning Axis Ticks And Labels

```python
review_cols = ["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]

movie = norm_reviews.loc[0, :]
print(movie)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar([1,2,3,4,5], movie[review_cols], width=-0.3, align="edge")

ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(review_cols, rotation=90)

ax.set_xlabel("Rating Source")
ax.set_ylabel("Average score")
ax.set_title("Average user rating for Avengers: Age of Ultron")
plt.show()
```

### Horizontal Bar Plot

```python
review_cols = ["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
movie = norm_reviews.loc[0, :]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.barh([1, 2, 3, 4, 5], movie[review_cols])

ax.set_xlabel("Average Rating")
ax.set_ylabel("Rating Source")
ax.set_title("Avengers: Age of Ultron")

plt.show()
```

```python
review_cols = ["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
movie = norm_reviews.loc[0, :]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.barh([1, 2, 3, 4, 5], movie[review_cols])

ax.set_xlabel("Average Rating")
ax.set_ylabel("Rating Source")
ax.set_title("Avengers: Age of Ultron")

ax.set_yticks([1, 2, 3, 4, 5])
ax.set_yticklabels(review_cols)

plt.show()
```
